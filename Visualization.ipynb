```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('newdemography.csv')
df
```

```python
sns.set_style("darkgrid")
```

```python
# Example plot using the "whitegrid" style
sns.scatterplot(data=df, x="Year", y="Female 0-4")
plt.title("Count baby female per year")
plt.show()
```

## Count baby male

```python
sns.set_style("whitegrid")
sns.scatterplot(data=df, x="Year", y="Male 0-4")
plt.title("Count baby female per year")
plt.show()
```

### Count Underage female per year

```python
sns.scatterplot(data=df, x="Year", y="Female 5-17")
plt.title("Count underage female per year")
plt.show()
```
