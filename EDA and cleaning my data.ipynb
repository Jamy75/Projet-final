## convert into numeric part 2 from (float to INT)

```python
import re

columns_to_convert = ['Female 0-4', 'Female 5-17', 'Male 0-4', 'Male 5-17']

# Remove non-numeric characters and convert columns to floats
df[columns_to_convert] = df[columns_to_convert].replace(r'[^0-9.]', '', regex=True).astype(float)
```

#### In this code, the replace() method is used with a regular expression pattern r'[^0-9.]' to remove any non-numeric characters (except for the decimal point '.'). This ensures that only valid numeric values remain in the columns.

```python
columns_to_convert = ['Female 0-4', 'Female 5-17', 'Male 0-4', 'Male 5-17']

# Replace non-finite values with NaN
df[columns_to_convert] = df[columns_to_convert].replace([np.inf, -np.inf, np.nan],0 )
’
df[columns_to_convert] = df[columns_to_convert].replace({'*': np.nan})



# Convert columns to integer
#df[columns_to_convert] = df[columns_to_convert].astype(int)
df[columns_to_convert] = df[columns_to_convert].astype('Int64')
```
